```python
entry_value = ''
class App:
    def __init__(self, root):
        #setting title
        root.title("GSI calculator")
        #setting window size
        width=600
        height=500
        screenwidth = root.winfo_screenwidth()
        screenheight = root.winfo_screenheight()
        alignstr = '%dx%d+%d+%d' % (width, height, (screenwidth - width) / 2, (screenheight - height) / 2)
        root.geometry(alignstr)
        root.resizable(width=False, height=False)

        self.GLineEdit_735 = tk.Entry(root)
        self.GLineEdit_735["borderwidth"] = "1px"
        ft = tkFont.Font(family='Times', size=10)
        self.GLineEdit_735["font"] = ft
        self.GLineEdit_735["fg"] = "#333333"
        self.GLineEdit_735["justify"] = "center"
        self.GLineEdit_735["text"] = "link"
        self.GLineEdit_735.place(x=310, y=60, width=70, height=25)

        GLabel_395=tk.Label(root)
        GLabel_395["bg"] = "#ff5722"
        ft = tkFont.Font(family='Times',size=18)
        GLabel_395["font"] = ft
        GLabel_395["fg"] = "#000000"
        GLabel_395["justify"] = "center"
        GLabel_395["text"] = "Image to Predict"
        GLabel_395.place(x=70,y=40,width=200,height=60)

        GButton_429=tk.Button(root)
        GButton_429["bg"] = "#f0f0f0"
        ft = tkFont.Font(family='Times',size=10)
        GButton_429["font"] = ft
        GButton_429["fg"] = "#000000"
        GButton_429["justify"] = "center"
        GButton_429["text"] = "Insert"
        GButton_429.place(x=440,y=60,width=70,height=25)
        GButton_429["command"] = self.GButton_429_command

        self.GLabel_207=tk.Label(root)
        ft = tkFont.Font(family='Times',size=15)
        self.GLabel_207["font"] = ft
        self.GLabel_207["fg"] = "#333333"
        self.GLabel_207["justify"] = "center"
        self.GLabel_207["text"] = "Prediction value"
        self.GLabel_207.place(x=320,y=260,width=200,height=25)

        self.GLineEdit_175=tk.Entry(root)
        self.GLineEdit_175["borderwidth"] = "1px"
        ft = tkFont.Font(family='Times',size=10)
        self.GLineEdit_175["font"] = ft
        self.GLineEdit_175["fg"] = "#333333"
        self.GLineEdit_175["justify"] = "center"
        self.GLineEdit_175["text"] = "Entry"
        self.GLineEdit_175.place(x=310,y=350,width=70,height=25)

        GLabel_30=tk.Label(root)
        GLabel_30["bg"] = "#01aaed"
        ft = tkFont.Font(family='Times',size=18)
        GLabel_30["font"] = ft
        GLabel_30["fg"] = "#333333"
        GLabel_30["justify"] = "center"
        GLabel_30["text"] = "Surface Condition"
        GLabel_30.place(x=70,y=330,width=200,height=60)

        self.GLabel_278=tk.Label(root)
        ft = tkFont.Font(family='Times',size=18)
        self.GLabel_278["font"] = ft
        self.GLabel_278["fg"] = "#333333"
        self.GLabel_278["justify"] = "center"
        self.GLabel_278["text"] = "Answer"
        self.GLabel_278.place(x=310,y=420,width=200,height=30)

        GLabel_939=tk.Label(root)
        GLabel_939["bg"] = "#fad400"
        ft = tkFont.Font(family='Times',size=18)
        GLabel_939["font"] = ft
        GLabel_939["fg"] = "#333333"
        GLabel_939["justify"] = "center"
        GLabel_939["text"] = "Predict with Picture"
        GLabel_939.place(x=70,y=140,width=200,height=60)

        GLabel_844=tk.Label(root)
        GLabel_844["bg"] = "#45f542"
        ft = tkFont.Font(family='Times',size=18)
        GLabel_844["font"] = ft
        GLabel_844["fg"] = "#333333"
        GLabel_844["justify"] = "center"
        GLabel_844["text"] = "Prediction Results"
        GLabel_844.place(x=70,y=250,width=200,height=60)

        GButton_818=tk.Button(root)
        GButton_818["bg"] = "#f0f0f0"
        ft = tkFont.Font(family='Times',size=10)
        GButton_818["font"] = ft
        GButton_818["fg"] = "#000000"
        GButton_818["justify"] = "center"
        GButton_818["text"] = "Enter"
        GButton_818.place(x=440,y=350,width=70,height=25)
        GButton_818["command"] = self.GButton_818_command

        self.GLabel_771=tk.Label(root)
        self.GLabel_771["bg"] = "#a442f5"
        ft = tkFont.Font(family='Times',size=18)
        self.GLabel_771["font"] = ft
        self.GLabel_771["fg"] = "#333333"
        self.GLabel_771["justify"] = "center"
        self.GLabel_771["text"] = "GSI Value"
        self.GLabel_771.place(x=70,y=410,width=200,height=60)

        self.pic = tk.Label(root)
        ft = tkFont.Font(family='Times', size=18)
        self.pic["font"] = ft
        self.pic["justify"] = "center"
        self.pic.place(x=310, y=100, width=140, height=140)

    def GButton_429_command(self):
        # Get the text from the Entry widget
        entry_value = self.GLineEdit_735.get()
        print(entry_value)
        img_path = entry_value
        img = load_img(img_path, target_size=(225, 225))
        img_array = img_to_array(img)
        # img_array = preprocess_input(img_array)
        img_array = img_array.reshape((1, 225, 225, 3))
        predictions = model.predict(img_array)
        answer = ''
        if predictions[0][0] == 1:
            answer = 'Intact or massive'
        elif predictions[0][1] == 1:
            answer = 'Blocky'
        elif predictions[0][2] == 1:
            answer = 'Very Blocky'
        elif predictions[0][3] == 1:
            answer = 'Blocky/Disturbed'
        elif predictions[0][4] == 1:
            answer = 'Disintegrated'
        self.GLabel_207.config(text=answer)
        image = Image.open(entry_value)
        image = image.resize((140, 140))
        photo = ImageTk.PhotoImage(image)
        self.pic.config(image=photo)
        self.pic.image = photo

    def GButton_818_command(self):
        entry_value = self.GLineEdit_735.get()
        # print(entry_value)
        img_path = entry_value
        img = load_img(img_path, target_size=(225, 225))
        img_array = img_to_array(img)
        # img_array = preprocess_input(img_array)
        img_array = img_array.reshape((1, 225, 225, 3))
        predictions = model.predict(img_array)
        answer = ''
        if predictions[0][0] == 1:
            answer = 'Intact or massive'
        elif predictions[0][1] == 1:
            answer = 'Blocky'
        elif predictions[0][2] == 1:
            answer = 'Very Blocky'
        elif predictions[0][3] == 1:
            answer = 'Blocky/Disturbed'
        elif predictions[0][4] == 1:    
            answer = 'Disintegrated'
        final = ''
        val = self.GLineEdit_175.get().lower()
        if answer == 'Intact or massive':
            if val == 'very good':
                final = '100-77'
            elif val == 'good':
                final ='89-67'
            elif val == 'fair':
                final ='80-56'
            elif val == 'poor':
                final ='NA'
            elif val == 'very poor':
                final ='NA'
        elif answer == 'Blocky':
            if val == 'very good':
                final ='86-65'
            elif val == 'good':
                final ='77-55'
            elif val == 'fair':
                final ='66-45'
            elif val == 'poor':
                final ='55-35'
            elif val == 'very poor':
                final ='43-26'
        elif answer == 'Very Blocky':
            if val == 'very good':
                final ='74-55'
            elif val == 'good':
                final ='65-45'
            elif val == 'fair':
                final ='54-36'
            elif val == 'poor':
                final ='44-27'
            elif val == 'very poor':
                final ='34-18'
        elif answer == 'Blocky/Disturbed':
            if val == 'very good':
                final ='63-45'
            elif val == 'good':
                final ='54-38'
            elif val == 'fair':
                final ='45-29'
            elif val == 'poor':
                final ='36-20'
            elif val == 'very poor':
                final ='27-11'
        elif answer == 'Disintegrated':
            if val == 'very good':
                final ='54-39'
            elif val == 'good':
                final ='46-30'
            elif val == 'fair':
                final ='37-23'
            elif val == 'poor':
                final ='29-15'
            elif val == 'very poor':
                final ='20-5'
        self.GLabel_278.config(text=final)

if __name__ == "__main__":
    root = tk.Tk()
    app = App(root)
    root.mainloop()
```
